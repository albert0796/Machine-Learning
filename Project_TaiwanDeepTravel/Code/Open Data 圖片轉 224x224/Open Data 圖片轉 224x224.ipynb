```python
import os
import glob
from PIL import Image
import os.path

file = "/Users/cloud/Desktop/open_data_image/*.jpg"   #須轉檔之位置
saveDir = "/Users/cloud/Desktop/open_data_image_resized"#轉檔後存放地點  
count = 0 
def convertjpg(jpgfile,saveDir,width=224,height=224):#可設定大小
    img=Image.open(jpgfile) 
    new_img=img.resize((width,height),Image.BILINEAR) #重新設定大小
    new_img.save(os.path.join(saveDir,os.path.basename(jpgfile)))#儲存
    


def modifyjpgSize(file,saveDir): 
    for jpgfile in glob.glob(file):
        convertjpg(jpgfile,saveDir)
        global count
        count += 1

modifyjpgSize(file,saveDir)
print("共完成",count,"件")
```
