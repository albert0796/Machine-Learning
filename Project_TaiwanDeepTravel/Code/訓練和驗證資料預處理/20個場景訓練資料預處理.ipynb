# 將資料從256x256 pixels轉224x224 pixels

```python
startTime = time()#開始計算時間

for i in range(100000):
    im = Image.open(image_filename_list[i])
    #print(im.size)
    width = 224
    ratio = float(width)/im.size[0]
    height = int(im.size[1]*ratio)
    nim = im.resize( (width, height), Image.BILINEAR )
    #print(nim.size)
    name2 = str(i+1)
    while(len(name2)<8):
        name2 = "0" + name2
    name3 = ".jpg"
    nim.save('C:/Users/firzen41616316/Desktop/taiwan_typical_10_resized/' +name2 + name3)

duration = time() - startTime
print('Finished time: ', duration)
```

# 將訓練資料image做正規化(除255.0)並以numpy(.npy)檔形式儲存

```python
count=1#從第00000001.jpg筆開始讀
print('Start--------------------')
for k in range(100):#100000筆，每次1000筆，分100次跑
    imgonehot = np.zeros(shape=(1000, 224, 224, 3))#存放訓練圖片的1000筆三維陣列
    startTime = time()#開始計算時間
    for m in range(1000):
        name2 = str(count)
        while(len(name2)<8):
            name2 = "0" + name2
        name3 = ".jpg"
        image = io.imread('C:/Users/firzen41616316/Desktop/taiwan_typical_10_resized/' +name2 + name3)#取得圖片值
        image = image.astype('float32')/255.0#將100筆圖片全部標準化(介於0~1之間)
        imgonehot[m] = image
        count+=1
    np.save('C:/Users/firzen41616316/Desktop/numpydataKeras/imgonehot_'+ str((k+1)*1000)+'.npy', imgonehot )
    duration = time() - startTime
    print('Finished ', (k+1)*1000, ' image, time: ', duration)
print('Finished All------------------')
```
