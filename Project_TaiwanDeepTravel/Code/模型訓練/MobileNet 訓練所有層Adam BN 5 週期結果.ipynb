```python
import keras
from keras.models import Model
from keras.layers.core import Flatten, Dense, Dropout, Activation
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
import numpy as np
from keras.layers.normalization import BatchNormalization
import functools
```

```python
net = keras.applications.mobilenet.MobileNet(include_top=False, weights='imagenet', input_tensor=None, input_shape=(224, 224, 3))
```

```python
x = net.output
x = Flatten()(x)
x = BatchNormalization()(x)
x = Dense(128)(x)
x = Activation('relu')(x)
x = Dropout(0.5)(x)
x = BatchNormalization()(x)
x = Dense(20)(x)
output_layer = Activation('softmax')(x)
net_final = Model(inputs=net.input, outputs=output_layer)
```

```python
count=0
for layer in net_final.layers:
    count+=1
print(count)
```

```python
print(net_final.summary())
```

```python
for layer in net_final.layers:
    layer.trainable = True
```

```python
net_final.compile(Adam(lr=.00002), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
V1 = np.load("C:/Users/firzen518/Desktop/numpydataKeras_20_new_val/imgonehot_new_val_2000.npy")
V2 = np.load("C:/Users/firzen518/Desktop/numpydataKeras_20_new_val/labelonehot_new_val_2000.npy")
```

```python
for i in range(5):#訓練5週期
    for j in range(100):#100,000張分100批訓練
        X = np.load( "C:/Users/firzen518/Desktop/numpydataKeras/imgonehot_"+str((j+1)*1000)+".npy" )
        Y = np.load( "C:/Users/firzen518/Desktop/numpydataKeras/labelonehot_"+str((j+1)*1000)+".npy" )
        net_final.fit(x = X, y = Y,
                      epochs = 1,#訓練週期
                      validation_data=(V1, V2),
                      verbose = 1)#顯示訓練過程，1为输出进度条记录，2为每个epoch输出一行记录
        print('Done training ', (j+1)*1000 ,' images in Epoch : ', i)
    print('Done training 100000 images, Epoch ', i ,'  -------------')
```
